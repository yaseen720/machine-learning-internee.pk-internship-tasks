## Step 2: Synthetic (Fake) Dataset Banate Hain

```python
# importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# generating 100 interns data
np.random.seed(42)  

# 100 rows data
n_samples = 100

data = {
    'Attendance_Percentage': np.random.uniform(60, 100, n_samples).round(1),
    'Quiz_Score': np.random.uniform(40, 100, n_samples).round(1),
    'Hours_Spent_On_Tasks': np.random.uniform(50, 300, n_samples).round(0),
    'Previous_Experience': np.random.choice([0, 1], n_samples, p=[0.6, 0.4]),  # 40% have experience
    'Tasks_Submitted_On_Time': np.random.randint(0, 11, n_samples)
}

# creating DataFrame 
df = pd.DataFrame(data)

# calucating Final_Performance_Score 
df['Final_Performance_Score'] = (
    0.3 * df['Attendance_Percentage'] +
    0.25 * df['Quiz_Score'] +
    0.2 * (df['Hours_Spent_On_Tasks'] / 3) +  # normalized
    10 * df['Previous_Experience'] +
    3 * df['Tasks_Submitted_On_Time']
) + np.random.normal(0, 5, n_samples)  

df['Final_Performance_Score'] = df['Final_Performance_Score'].clip(0, 100).round(1)

print(df.head(10))
print(df.describe())
```

```python
print(df.corr()['Final_Performance_Score'].sort_values(ascending=False))

plt.figure(figsize=(8,6))
plt.scatter(df['Quiz_Score'], df['Final_Performance_Score'], alpha=0.7)
plt.title('Quiz Score vs Final Performance Score')
plt.xlabel('Quiz Score')
plt.ylabel('Final Performance Score')
plt.grid(True)
plt.show()
```

### Step 4: Model Training  (Linear Regression)

##### Predictions on test data

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# creating model
model = LinearRegression()

# model train (fit) on Training data 
model.fit(X_train, y_train)

# training completed
print("Model training complete ho gaya!")
```

```python
y_pred = model.predict(X_test)

# comparing 10 actual vs predicted values 
comparison = pd.DataFrame({
    'Actual_Score': y_test.values,
    'Predicted_Score': y_pred.round(1)
})

print("Actual vs Predicted (pehle 10):")
print(comparison.head(10))
```

##### Model Evaluation

```python
# calculating Metrics 
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Model Performance:")
print(f"Mean Absolute Error (MAE): {mae:.2f}   How much Average mistake  (in points)")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}    more penalty on big mistakes")
print(f"R Score: {r2:.3f}   1.0 perfect, 0.0 means model not good")

# Explanation:
# MAE ~5 means mistake of average 5 points
# R > 0.8 very good, 0.6-0.8 decent, <0.5 weak
```

#### Step 5: Trying Better Model  (Random Forest Regressor)

```python
from sklearn.ensemble import RandomForestRegressor

# creating new model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train (direct without scaling)
rf_model.fit(X_train, y_train)

# Predictions
y_pred_rf = rf_model.predict(X_test)

# Metrics
mae_rf = mean_absolute_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
r2_rf = r2_score(y_test, y_pred_rf)

print("Random Forest Model Performance:")
print(f"MAE: {mae_rf:.2f}")
print(f"RMSE: {rmse_rf:.2f}")
print(f"R Score: {r2_rf:.3f}")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline  # Easy way to combine steps

# Pipeline: Scaling + Linear Regression
final_model = Pipeline([
    ('scaler', StandardScaler()),
    ('linear', LinearRegression())
])

# Train
final_model.fit(X_train, y_train)

# Predictions and metrics
y_pred_final = final_model.predict(X_test)

mae_final = mean_absolute_error(y_test, y_pred_final)
rmse_final = np.sqrt(mean_squared_error(y_test, y_pred_final))
r2_final = r2_score(y_test, y_pred_final)

print("Final Model (Linear Regression with Scaling):")
print(f"MAE: {mae_final:.2f}")
print(f"RMSE: {rmse_final:.2f}")
print(f"R Score: {r2_final:.3f}")
```

##### Adding Polynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures

# Pipeline with Polynomial (degree=2)
poly_model = Pipeline([
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('scaler', StandardScaler()),
    ('linear', LinearRegression())
])

poly_model.fit(X_train, y_train)

y_pred_poly = poly_model.predict(X_test)

mae_poly = mean_absolute_error(y_test, y_pred_poly)
rmse_poly = np.sqrt(mean_squared_error(y_test, y_pred_poly))
r2_poly = r2_score(y_test, y_pred_poly)

print("\nWith Polynomial Features (degree=2):")
print(f"MAE: {mae_poly:.2f}")
print(f"RMSE: {rmse_poly:.2f}")
print(f"R Score: {r2_poly:.3f}")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
import joblib

# Final pipeline (best version)
final_intern_model = Pipeline([
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('scaler', StandardScaler()),
    ('linear', LinearRegression())
])

final_intern_model.fit(X_train, y_train)

y_pred_final = final_intern_model.predict(X_test)
print("Final Confirmed Performance:")
print(f"R Score: {r2_score(y_test, y_pred_final):.3f}")
print(f"MAE: {mean_absolute_error(y_test, y_pred_final):.2f}")
```

```python
# Model save 
joblib.dump(final_intern_model, 'intern_performance_prediction_model.pkl')

print("Final model saved as 'intern_performance_prediction_model.pkl'")
```
