### Create Sample FAQ Dataset

```python
faq_data = {
    "question": [
        "How do I apply for an internship?",
        "What is the duration of the internship?",
        "Will I get a certificate after completion?",
        "How do I submit my tasks?",
        "Is this internship paid?",
        "Who can I contact for technical issues?",
        "How are interns evaluated?",
        "Can I work remotely?"
    ],
    "answer": [
        "You can apply through the Internee.pk official website.",
        "The internship duration is typically 1 to 3 months.",
        "Yes, interns receive a certificate after successful completion.",
        "Tasks must be submitted through the assigned platform or GitHub.",
        "This internship is unpaid but provides valuable experience.",
        "You can contact the support team via the dashboard or email.",
        "Interns are evaluated based on task completion and performance.",
        "Yes, the internship is fully remote."
    ]
}

faq_df = pd.DataFrame(faq_data)
faq_df
```

### Save Dataset

```python
faq_df.to_csv("internship_faq.csv", index=False)
print("FAQ dataset saved as internship_faq.csv")
```

## Text Preprocessing for Chatbot

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download("punkt")
nltk.download("stopwords")
```

### Text Cleaning Function

```python
stop_words = set(stopwords.words("english"))

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", "", text)
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    return " ".join(tokens)
```

```python
import nltk
nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("stopwords")
```

### Apply Preprocessing

```python
faq_df["clean_question"] = faq_df["question"].apply(preprocess_text)
faq_df["clean_answer"] = faq_df["answer"].apply(preprocess_text)

faq_df
```

```python
faq_df.to_csv("internship_faq_cleaned.csv", index=False)
print("Cleaned FAQ dataset saved successfully")
```

### Word Count Visualization

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

### Loading Pretrained Model

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
question_embeddings = model.encode(
    faq_df["clean_question"].tolist(),
    convert_to_numpy=True
)

question_embeddings.shape
```

```python
np.save("faq_question_embeddings.npy", question_embeddings)
print("Embeddings saved successfully")
```

### Similarity Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns

similarity_matrix = cosine_similarity(question_embeddings)

plt.figure(figsize=(7,6))
sns.heatmap(similarity_matrix, cmap="coolwarm")
plt.title("FAQ Question Similarity Heatmap")
plt.tight_layout()
plt.savefig("faq_similarity_heatmap.png")
plt.show()
```

## Define Chatbot Function

```python
def chatbot_response(user_query, faq_df, embeddings, model):
    # preprocess user input
    clean_query = preprocess_text(user_query)
    
    # encode user query
    query_embedding = model.encode([clean_query], convert_to_numpy=True)
    
    # compute similarity
    similarities = cosine_similarity(query_embedding, embeddings)
    
    # find best match
    best_match_idx = np.argmax(similarities)
    
    return faq_df.iloc[best_match_idx]["answer"]
```

### Testing Chatbot

```python
test_query = "Will I receive a certificate?"
response = chatbot_response(test_query, faq_df, question_embeddings, model)

print("User:", test_query)
print("Chatbot:", response)
```

### Test Multiple Queries

```python
test_questions = [
    "Is this internship remote?",
    "How long is the internship?",
    "Where do I submit my tasks?",
    "Is this internship paid?"
]

for q in test_questions:
    print("Q:", q)
    print("A:", chatbot_response(q, faq_df, question_embeddings, model))
    print("-" * 40)
```

### Save Chatbot Demo Results

```python
demo_results = []

for q in test_questions:
    demo_results.append({
        "User Question": q,
        "Chatbot Answer": chatbot_response(q, faq_df, question_embeddings, model)
    })

demo_df = pd.DataFrame(demo_results)
demo_df.to_csv("chatbot_demo_results.csv", index=False)

demo_df
```
