# Dataset Creation & Problem Setup

### Import Required Libraries

```python
import pandas as pd
import numpy as np
```

### Create Intern Skills Dataset


```python
intern_data = {
    "intern_id": [1, 2, 3, 4],
    "intern_role": [
        "Machine Learning Intern",
        "Web Development Intern",
        "Data Analyst Intern",
        "AI Intern"
    ],
    "skills": [
        "Python, Pandas, Machine Learning, Scikit-learn",
        "HTML, CSS, JavaScript",
        "SQL, Excel, Data Visualization",
        "Python, Deep Learning, NLP"
    ]
}

intern_df = pd.DataFrame(intern_data)
intern_df
```

### Create Industry Job Requirement Dataset

```python
job_data = {
    "job_role": [
        "Machine Learning Engineer",
        "Frontend Developer",
        "Data Analyst",
        "AI Engineer"
    ],
    "required_skills": [
        "Python, Machine Learning, Deep Learning, TensorFlow, SQL",
        "HTML, CSS, JavaScript, React",
        "SQL, Excel, Python, Data Visualization",
        "Python, Deep Learning, NLP, Transformers"
    ]
}

job_df = pd.DataFrame(job_data)
job_df
```

### Combine Intern & Job Skill Text (For NLP)

```python
intern_df["text"] = intern_df["skills"]
job_df["text"] = job_df["required_skills"]

intern_df, job_df
```

# TF-IDF Vectorization (Skill Text → Numbers)

### Import TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

### Combine Intern + Job Skill Text

```python
combined_text = pd.concat([
    intern_df["text"],
    job_df["text"]
], axis=0)
```

### Apply TF-IDF

```python
tfidf = TfidfVectorizer(
    stop_words="english"
)

tfidf_matrix = tfidf.fit_transform(combined_text)

tfidf_matrix
```

### Convert TF-IDF to DataFrame (Optional but Useful)

```python
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf.get_feature_names_out()
)

tfidf_df.head()
```

# Skill Clustering using K-Means

### Import K-Means

```python
from sklearn.cluster import KMeans
```

### Decide Number of Clusters (k)

```python
kmeans = KMeans(
    n_clusters=3,
    random_state=42
)

clusters = kmeans.fit_predict(tfidf_matrix)
clusters
```

### Add Cluster Labels to DataFrames

```python
intern_df["cluster"] = clusters[:len(intern_df)]
job_df["cluster"] = clusters[len(intern_df):]

intern_df, job_df
```

### Visualization – Skill Clusters

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(tfidf_matrix.toarray())

plt.figure(figsize=(7,5))
plt.scatter(
    reduced_data[:len(intern_df), 0],
    reduced_data[:len(intern_df), 1],
    label="Intern Skills",
    marker="o"
)

plt.scatter(
    reduced_data[len(intern_df):, 0],
    reduced_data[len(intern_df):, 1],
    label="Industry Skills",
    marker="^"
)

plt.title("Skill Clustering: Interns vs Industry")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend()

plt.savefig(
    "skill_clusters_pca.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

# Skill Gap Identification & Training Recommendation

### Create Skill Sets

```python
def extract_skills(skill_text):
    return set([skill.strip().lower() for skill in skill_text.split(",")])
```

### Identify Skill Gaps for Each Intern

```python
gap_analysis = []

for _, intern in intern_df.iterrows():
    intern_skills = extract_skills(intern["skills"])

    # Match intern role with closest industry role using cluster
    matched_jobs = job_df[job_df["cluster"] == intern["cluster"]]

    if len(matched_jobs) > 0:
        industry_skills = extract_skills(matched_jobs.iloc[0]["required_skills"])
    else:
        industry_skills = set()

    missing_skills = industry_skills - intern_skills

    gap_analysis.append({
        "Intern ID": intern["intern_id"],
        "Intern Role": intern["intern_role"],
        "Current Skills": ", ".join(intern_skills),
        "Missing Skills": ", ".join(missing_skills)
    })

gap_df = pd.DataFrame(gap_analysis)
gap_df
```

### Training Recommendation Logic

```python
gap_df["Training Recommendation"] = gap_df["Missing Skills"].apply(
    lambda x: "No major gaps" if x == "" else "Recommended to learn: " + x
)

gap_df
```

### Visualization – Skill Gaps per Intern

```python
gap_df["Gap Count"] = gap_df["Missing Skills"].apply(
    lambda x: 0 if x == "" else len(x.split(","))
)

plt.figure(figsize=(6,4))
plt.bar(
    gap_df["Intern Role"],
    gap_df["Gap Count"]
)
plt.title("Number of Skill Gaps per Intern")
plt.ylabel("Skill Gap Count")
plt.xticks(rotation=30)

plt.savefig(
    "skill_gap_count_per_intern.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

### Save Skill Gap Analysis Results

```python
gap_df.to_csv("skill_gap_analysis.csv", index=False)
gap_df.to_excel("skill_gap_analysis.xlsx", index=False)
```
