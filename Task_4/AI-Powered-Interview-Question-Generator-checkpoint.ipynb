### Problem Setup & Sample Dataset Creation

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "intern_role": [
        "Machine Learning Intern",
        "Web Development Intern",
        "Data Analyst Intern"
    ],
    "skills": [
        "Python, Machine Learning, Pandas, Scikit-learn",
        "HTML, CSS, JavaScript, React",
        "SQL, Excel, Data Visualization"
    ],
    "job_description": [
        "Build and evaluate machine learning models for prediction tasks",
        "Develop responsive websites and web applications",
        "Analyze datasets and generate insights for decision making"
    ]
}

df = pd.DataFrame(data)
df
```

```python
df["prompt_text"] = (
    "Role: " + df["intern_role"] +
    ". Skills: " + df["skills"] +
    ". Job Description: " + df["job_description"]
)

df[["intern_role", "prompt_text"]]
```

### Load Text Generation Model (Transformer)

```python
!pip install transformers torch
```

```python
from transformers import pipeline
```

```python
generator = pipeline(
    "text-generation",
    model="gpt2",
    device=-1  # CPU
)
```

### Generate Technical Interview Questions

```python
prompt = (
    df.loc[0, "prompt_text"] +
    ". Generate 5 technical interview questions."
)

output = generator(
    prompt,
    max_length=150,
    num_return_sequences=1
)

print(output[0]["generated_text"])
```

#### Generate Questions for All Roles

```python
all_questions = {}

for idx, row in df.iterrows():
    prompt = row["prompt_text"] + ". Generate 5 technical interview questions."
    output = generator(
        prompt,
        max_length=150,
        truncation=True,  # avoid warnings
        num_return_sequences=1
    )
    generated_text = output[0]["generated_text"]
    all_questions[row["intern_role"]] = generated_text

all_questions
```

```python
prompt + ". Generate 3 behavioral interview questions."
```

```python
# Convert dictionary to DataFrame
questions_df = pd.DataFrame(list(all_questions.items()), columns=["Intern Role", "Generated Questions"])

# Save as CSV
questions_df.to_csv("generated_interview_questions.csv", index=False)

# Save as Excel
questions_df.to_excel("generated_interview_questions.xlsx", index=False)

questions_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

questions_df["Question Length"] = questions_df["Generated Questions"].apply(len)

plt.figure(figsize=(6,4))
sns.barplot(x="Intern Role", y="Question Length", data=questions_df)
plt.title("Length of Generated Interview Questions per Role")
plt.ylabel("Number of Characters")
plt.xticks(rotation=30)

plt.savefig("questions_length_per_role.png", dpi=300, bbox_inches="tight")
plt.show()
```
